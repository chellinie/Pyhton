```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

X = df["new"]
feature_model = CountVectorizer()
Y_new = feature_model.fit_transform(X)
Y_new_array = Y_new.toarray()
first_row = Y_new_array[9]

print(np.array2string(first_row, max_line_width=np.inf))
```

```python
feature_model.vocabulary_
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(n_neighbors=10)
model.fit(Y_new)
```

```python
cerita = ['historical movie', 'music', 'drama tv', 'kids ova', 'romance tv']
X_test = feature_model.transform(cerita)
dist, recommendation = model.kneighbors(X_test, n_neighbors=10)
```

```python
recommendation
```

```python
df.iloc[recommendation[4]]
```

```python
predicted = recommendation[4]
actual = [281, -1, -1, 160, 139, 137,  28, 299, 403,  -1]
```

```python
from sklearn.metrics import recall_score
score = recall_score(actual, predicted, average='macro')
```

```python
print(score)
```
