```python
print(f"BLEU Score ChatGPT: {bleu_chatgptI}")
print(f"BLEU Score Gemini: {bleu_geminiI}")
```

```python
# Menampilkan perbandingan BLEU score menggunakan grafik
labels = ['ChatGPT', 'Gemini']
scores = [bleu_chatgptI, bleu_geminiI]
```

```python
plt.bar(labels, scores, color=['blue', 'green'])
plt.ylabel('BLEU Score')
plt.title('BLEU Score ChatGPT vs Gemini Terhadap Prompt Mengenai Manajemen Sumber Daya Manusia')
plt.show()
```
